```python
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline  

cwd = os.getcwd()
```

```python
f = open("../data/Two_Patterns_TRAIN", "r")

x_label = []
x_feature = []

for i, line in enumerate(f):
    line = line.strip().split(" ")
    line = list(filter(lambda x: len(x) > 0, line))
    
    x_label.append(float(line[0]))
    x_feature.append(line[1:])
    
x_label = np.array(x_label, np.int32)
x_feature = np.array(x_feature, np.float)
```

```python
def plot_class(ax, clazz):
    x_idx = x_label == clazz
    ax.set(title='Class ' + str(clazz))
    ax.grid()
    for x in x_feature[x_idx]:
        ax.plot(x, 'C' + str(clazz % 10))
```

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
for i in range(3):
    for j in range(3):
        plot_class(axes[i, j] ,i*3+j+1)
plt.show()
```
